# Vectors — practice (NumPy re-derivation)

Worked solutions, executed. Read `theory.html` first; try each yourself before peeking.

```python
import numpy as np

a = np.array([4, 2], dtype=float)
b = np.array([1, 3], dtype=float)
print('a+b =', a + b)
print('2a  =', 2*a)
print('|a| =', np.linalg.norm(a))
```

### E1 (easy) — addition / scalar / subtraction
`a=[2,-1,4]`, `b=[0,3,1]`.

```python
a1=np.array([2,-1,4.]); b1=np.array([0,3,1.])
print('a+b =', a1+b1)
print('3a  =', 3*a1)
print('a-b =', a1-b1)
```

### E2 (easy) — magnitude
`||[6,8]||` by hand = 10 (a 6-8-10 triple).

```python
print(np.linalg.norm([6,8]))
```

### E3 (medium) — safe normalize

```python
def unit(v):
    v=np.asarray(v,float); n=np.linalg.norm(v)
    return v if n==0 else v/n
print(unit([3,4]))
print(unit([0,0]))
```

### E4 (medium) — feature scaling flips 'closest'

```python
h1=np.array([1200,3.]); h2=np.array([1400,2.])
print('raw   dist:', np.linalg.norm(h1-h2))
s1=np.array([1.2,3.]); s2=np.array([1.4,2.])
print('scaled dist:', np.linalg.norm(s1-s2))
```

### E5 (hard) — magnitude via dot product
||v|| = sqrt(v·v).

```python
rng=np.random.default_rng(0); v=rng.standard_normal(1000)
mag=np.sqrt(np.dot(v,v))
print(mag, np.allclose(mag, np.linalg.norm(v)))
```
