# Span, basis, rank & independence — practice (worked & executed)

```python
import numpy as np
print("indep rank", np.linalg.matrix_rank([[1,0],[0,1]]))
print("dep rank", np.linalg.matrix_rank([[1,2],[2,4]]), "det", np.linalg.det([[1,2],[2,4]]))
```

### E1 — independence by rank

```python
print(np.linalg.matrix_rank([[1,2],[2,4]]), np.linalg.matrix_rank([[1,2],[2,1]]))  # 1, 2
```

### E2 — 3 vectors in R^2 (max independent = dim)

```python
print("rank of 3 cols in R^2:", np.linalg.matrix_rank([[1,0,2],[0,1,3.]]))  # <= 2
```

### E3 — engineered dependence (c3 = c1 + c2)

```python
A=np.array([[1,0,1],[0,1,1],[0,0,0.]]); print(np.linalg.matrix_rank(A), round(np.linalg.det(A),6))
```

### E4 — dummy-variable trap

```python
oh=np.array([[1,0,0],[0,1,0],[0,0,1],[1,0,0.]]); X=np.column_stack([oh,np.ones(4)])
print("cols",X.shape[1],"rank",np.linalg.matrix_rank(X))  # 4 cols, rank 3
```

### E5 — keep vs break independence

```python
keep=np.array([[1,1,0],[0,1,1],[0,0,1.]]).T
brk =np.array([[1,1,0],[0,1,1],[1,2,1.]]).T
print("keep", np.linalg.matrix_rank(keep), "| break", np.linalg.matrix_rank(brk))  # 3, 2
```
