```python
import numpy as np

a = np.array([4.0, 2.0]); b = np.array([1.0, 3.0])
print("a . b   =", np.dot(a, b))
print("a @ b   =", a @ b)
print("a . a   =", a @ a, " |a|^2 =", np.linalg.norm(a)**2)

def cosine(a, b):
    na, nb = np.linalg.norm(a), np.linalg.norm(b)
    return 0.0 if na == 0 or nb == 0 else float(np.clip(np.dot(a, b)/(na*nb), -1, 1))
print("cos(a,b) =", cosine(a, b))
```
