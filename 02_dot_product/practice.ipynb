# Dot product & similarity — practice (worked & executed)

```python
import numpy as np
a=np.array([4,2.]); b=np.array([1,3.])
print("a.b =", np.dot(a,b), "| a@b =", a@b)
print("a.a, |a|^2 =", a@a, np.linalg.norm(a)**2)
def cosine(a,b):
    a=np.asarray(a,float); b=np.asarray(b,float)
    na,nb=np.linalg.norm(a),np.linalg.norm(b)
    return 0.0 if na==0 or nb==0 else float(np.clip(np.dot(a,b)/(na*nb),-1,1))
print("cos(a,b) =", round(cosine(a,b),4))
```

### E3 — safe cosine

```python
print(cosine([1,2],[2,4]), cosine([0,0],[1,1]))
```

### E4 — rank docs by cosine

```python
q=np.array([1,0,1.])
for k,d in {"d1":[1,0,1],"d2":[2,0,2],"d3":[0,1,0]}.items():
    print(k, round(cosine(q,d),3))   # d1,d2 tie at 1.0; d3=0
```

### E5 — projection vector

```python
a=np.array([3,4.]); b=np.array([1,0.])
print((np.dot(a,b)/np.dot(b,b))*b)   # [3. 0.]
```
