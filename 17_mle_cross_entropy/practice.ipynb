# MLE & cross-entropy — practice (worked & executed)

```python
import numpy as np
k,n=7,10; ps=np.linspace(0.01,0.99,99)
ll=[k*np.log(p)+(n-k)*np.log(1-p) for p in ps]
print("MLE", round(ps[int(np.argmax(ll))],2))   # 0.70
y=np.array([1,0,1,1]); q=np.array([0.9,0.2,0.8,0.6])
print("BCE", round(-np.mean(y*np.log(q)+(1-y)*np.log(1-q)),4))
```

### E1 — coin MLE

```python
print(round(ps[int(np.argmax(ll))],2))   # 0.70
```

### E2 — binary cross-entropy

```python
y=np.array([1,0,1]); q=np.array([0.9,0.1,0.8])
print(round(-np.mean(y*np.log(q)+(1-y)*np.log(1-q)),4))
```

### E3 — Gaussian MLE = sample mean

```python
x=np.array([2.,4,4,5,9]); mus=np.linspace(0,10,1001)
print(round(mus[int(np.argmin([((x-m)**2).sum() for m in mus]))],3), round(x.mean(),3))
```

### E4 — cross-entropy near 0

```python
print([round(-np.log(qc),3) for qc in (0.5,0.1,0.01)])   # rises: 0.693, 2.303, 4.605
```

### E5 — ridge = L2 prior

```python
rng=np.random.default_rng(0); x=rng.random(50); y=3*x+0.1*rng.standard_normal(50)
for lam in [0,1,10,100]:
    w=(x@y)/(x@x+lam); print(lam, round(w,3))   # w shrinks toward 0
```
