# MLE & cross-entropy — practice

Read `theory.html` first. Every loss is a negative log-likelihood.

```python
import numpy as np

# coin MLE = observed proportion
k, n = 7, 10
ps = np.linspace(0.01, 0.99, 99)
ll = [k*np.log(p) + (n-k)*np.log(1-p) for p in ps]
print('MLE:', round(ps[int(np.argmax(ll))], 2))   # 0.70

# binary cross-entropy = NLL of Bernoulli labels
y = np.array([1,0,1,1]); q = np.array([0.9,0.2,0.8,0.6])
print('BCE:', round(-np.mean(y*np.log(q)+(1-y)*np.log(1-q)), 4))
```
