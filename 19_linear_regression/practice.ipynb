# Linear regression & bias–variance — practice

Read `theory.html` first. Least squares = Gaussian MLE; always check residuals + test error.

```python
import numpy as np
rng = np.random.default_rng(0)
n = 100
x = rng.uniform(0,5,n)
y = 2.5*x + 1.0 + rng.normal(0,1.5,n)
X = np.column_stack([np.ones(n), x])
w = np.linalg.lstsq(X, y, rcond=None)[0]
print('w:', w.round(3))                     # [0.989 2.461]
yhat = X @ w
print('R2:', round(1 - np.sum((y-yhat)**2)/np.sum((y-y.mean())**2), 3))   # 0.868
print('normal eqn matches:', np.allclose(w, np.linalg.inv(X.T@X)@X.T@y))
```
