# Linear regression & bias–variance — practice (worked & executed)

```python
import numpy as np
rng=np.random.default_rng(0); n=100
x=rng.uniform(0,5,n); y=2.5*x+1.0+rng.normal(0,1.5,n)
X=np.column_stack([np.ones(n),x]); w=np.linalg.lstsq(X,y,rcond=None)[0]
yhat=X@w
print("w", w.round(3), "| R2", round(1-np.sum((y-yhat)**2)/np.sum((y-y.mean())**2),3))
print("normal eqn matches:", np.allclose(w, np.linalg.inv(X.T@X)@X.T@y))
```

### E1 — recover the line

```python
print(np.linalg.lstsq(X,y,rcond=None)[0].round(3))   # ~[1.0, 2.5]
```

### E2 — R^2 from residuals

```python
print(round(1-np.sum((y-yhat)**2)/np.sum((y-y.mean())**2),3))
```

### E3 — collinearity

```python
Xc=np.column_stack([np.ones(n), x, x])   # duplicated feature
print("rank", np.linalg.matrix_rank(Xc), "of", Xc.shape[1], "cols -> singular X^T X")
print("lstsq still returns a solution:", np.linalg.lstsq(Xc,y,rcond=None)[0].round(2))
```

### E4 — bias-variance table

```python
xt=np.sort(rng.uniform(0,1,60)); yt=np.sin(2*np.pi*xt)+rng.normal(0,0.25,60)
xtr,ytr,xte,yte=xt[:40],yt[:40],xt[40:],yt[40:]
for d in [1,3,9,15]:
    c=np.polyfit(xtr,ytr,d)
    print(d, round(np.mean((np.polyval(c,xtr)-ytr)**2),3), round(np.mean((np.polyval(c,xte)-yte)**2),3))
```

### E5 — ridge dial

```python
def ridge(X,y,lam): return np.linalg.solve(X.T@X+lam*np.eye(X.shape[1]), X.T@y)
for lam in [0,1,10,100]: print(lam, np.round(ridge(X,y,lam),3))
```
