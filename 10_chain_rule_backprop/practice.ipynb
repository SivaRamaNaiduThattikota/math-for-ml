# Chain rule & backprop — practice

Read `theory.html` first. Derive by hand, then check with finite differences.

```python
import numpy as np
sig = lambda z: 1/(1+np.exp(-z))

x, y, w = 1.5, 1.0, 0.5
# forward
z = w*x; a = sig(z); L = (a-y)**2
# backward (chain rule, right to left)
dL_da = 2*(a-y); dL_dz = dL_da*a*(1-a); dL_dw = dL_dz*x
print('L =', L, '  dL/dw =', dL_dw)

g = lambda w: (sig(w*x)-y)**2
print('finite diff:', (g(w+1e-6)-g(w-1e-6))/2e-6)   # matches dL_dw
```
