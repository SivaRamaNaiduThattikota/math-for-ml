# Chain rule & backprop — practice (worked & executed)

```python
import numpy as np
sig=lambda z:1/(1+np.exp(-z)); x,y,w=1.5,1.0,0.5
z=w*x; a=sig(z); L=(a-y)**2
dL_dw=2*(a-y)*a*(1-a)*x
g=lambda w:(sig(w*x)-y)**2
print("L",round(L,4),"dL/dw",round(dL_dw,4),"finite",round((g(w+1e-6)-g(w-1e-6))/2e-6,4))
```

### E1 — chain rule by hand (numeric check)

```python
d=lambda f,x,h=1e-6:(f(x+h)-f(x-h))/(2*h)
print(round(d(lambda x:np.sin(x**2),1.3),4), round(2*1.3*np.cos(1.3**2),4))
print(round(d(lambda x:np.exp(3*x),0.5),4), round(3*np.exp(1.5),4))
```

### E2 — dL/dz

```python
print("dL/dz =", round(2*(a-y)*a*(1-a),4))
```

### E3 — single-neuron backprop matches finite diff

```python
print(np.isclose(dL_dw,(g(w+1e-6)-g(w-1e-6))/2e-6))
```

### E4 — two-layer chain dL/dw1

```python
x1,y1,w1,w2=1.0,2.0,0.5,0.8
z1=w1*x1; a1=sig(z1); z2=w2*a1
dL_dw1=2*(z2-y1)*w2*a1*(1-a1)*x1
gg=lambda w1:((w2*sig(w1*x1))-y1)**2
print(round(dL_dw1,5), round((gg(w1+1e-6)-gg(w1-1e-6))/2e-6,5))
```
