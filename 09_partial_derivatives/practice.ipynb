# Partial derivatives & the Jacobian — practice (worked & executed)

```python
import numpy as np
def grad(f,p,h=1e-5):
    p=np.asarray(p,float); g=np.zeros_like(p)
    for i in range(len(p)):
        s=np.zeros_like(p); s[i]=h; g[i]=(f(p+s)-f(p-s))/(2*h)
    return g
print(grad(lambda p:p[0]**2+p[1]**2,[3.,4.]), np.linalg.norm(grad(lambda p:p[0]**2+p[1]**2,[3.,4.])))
```

### E1 — partials by hand

```python
print(grad(lambda p:3*p[0]**2+p[0]*p[1]+p[1]**3,[1.,2.]))   # [8,13]
```

### E2 — where is grad zero

```python
print(grad(lambda p:p[0]**2+p[1]**2,[0.,0.]))   # [0,0]
```

### E3 — numerical gradient check

```python
print(grad(lambda p:p[0]**2+p[1]**2,[3.,4.]))   # [6,8]
```

### E4 — directional derivative

```python
g=grad(lambda p:p[0]**2+p[1]**2,[1.,1.]); u=np.array([1,0.]); ghat=g/np.linalg.norm(g)
print("along [1,0]:", g@u, "| along gradient:", round(g@ghat,3))
```

### E5 — Jacobian by hand + numeric column

```python
J=lambda x,y: np.array([[2*x*y, x*x],[1, np.cos(y)]])
print("J(1,0)=\n", J(1,0))
h=1e-5; F=lambda x,y:np.array([x*x*y,x+np.sin(y)])
print("numeric d/dx at (1,0):", np.round((F(1+h,0)-F(1-h,0))/(2*h),4))
```
