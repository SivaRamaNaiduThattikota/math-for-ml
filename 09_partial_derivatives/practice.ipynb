# Partial derivatives & the Jacobian — practice

Read `theory.html` first. Do partials by hand, then verify numerically.

```python
import numpy as np

def grad(f, p, h=1e-5):
    p = np.asarray(p, float); g = np.zeros_like(p)
    for i in range(len(p)):
        s = np.zeros_like(p); s[i] = h
        g[i] = (f(p+s) - f(p-s)) / (2*h)
    return g

print(grad(lambda p: p[0]**2 + p[1]**2, [3., 4.]))   # [6. 8.]
print(np.linalg.norm(grad(lambda p: p[0]**2 + p[1]**2, [3., 4.])))  # 10.0
```
