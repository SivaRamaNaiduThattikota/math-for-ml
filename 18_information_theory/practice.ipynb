# Information theory — practice (worked & executed)

```python
import numpy as np
entropy=lambda p:-np.sum(p*np.log2(p+1e-12))
cross_entropy=lambda p,q:-np.sum(p*np.log2(q+1e-12))
kl=lambda p,q:np.sum(p*np.log2((p+1e-12)/(q+1e-12)))
p=np.array([0.7,0.2,0.1]); q=np.array([0.5,0.3,0.2])
print(round(entropy(p),3), round(cross_entropy(p,q),3), round(kl(p,q),3))
print("CE=H+KL:", np.isclose(cross_entropy(p,q), entropy(p)+kl(p,q)))
```

### E1 — entropy of fair coin & d8

```python
print(entropy(np.array([0.5,0.5])), entropy(np.ones(8)/8))   # 1.0, 3.0
```

### E2 — binary entropy shape

```python
H=lambda p:-(p*np.log2(p)+(1-p)*np.log2(1-p)); print(round(H(0.5),3), round(H(0.9),3))
```

### E3 — CE = H + KL

```python
print(np.isclose(cross_entropy(p,q), entropy(p)+kl(p,q)))
```

### E4 — KL asymmetry

```python
print(round(kl(p,q),4), round(kl(q,p),4))   # differ
```

### E5 — one-hot label

```python
ph=np.array([0,1,0.]); qh=np.array([0.1,0.7,0.2])
print("entropy", entropy(ph), "| CE", round(cross_entropy(ph,qh),3), "| -log2 q(true)", round(-np.log2(0.7),3))
```
