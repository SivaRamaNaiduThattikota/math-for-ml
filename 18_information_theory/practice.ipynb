# Information theory — practice (final session)

Read `theory.html` first. Cross-entropy = entropy + KL.

```python
import numpy as np
entropy       = lambda p: -np.sum(p*np.log2(p+1e-12))
cross_entropy = lambda p,q: -np.sum(p*np.log2(q+1e-12))
kl            = lambda p,q: np.sum(p*np.log2((p+1e-12)/(q+1e-12)))

p = np.array([0.7,0.2,0.1]); q = np.array([0.5,0.3,0.2])
print(round(entropy(p),3), round(cross_entropy(p,q),3), round(kl(p,q),3))
print(np.isclose(cross_entropy(p,q), entropy(p)+kl(p,q)))   # True
```
