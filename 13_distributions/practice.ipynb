Read `theory.html` first. Sample, then check the summary numbers.

```python
import numpy as np

data = np.random.normal(loc=2.0, scale=0.5, size=10000)
print('mean ~', round(data.mean(),2), ' std ~', round(data.std(),2))

def normal_pdf(x, mu=0, sig=1):
    return np.exp(-(x-mu)**2/(2*sig**2)) / (sig*np.sqrt(2*np.pi))
print('peak of standard normal:', round(normal_pdf(0), 4))   # 0.3989
```
