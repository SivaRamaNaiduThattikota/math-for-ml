# Distributions — practice (worked & executed)

```python
import numpy as np
def normal_pdf(x,mu=0,sig=1): return np.exp(-(x-mu)**2/(2*sig**2))/(sig*np.sqrt(2*np.pi))
print("peak standard normal:", round(normal_pdf(0),4))   # 0.3989
```

### E1 — sample N(5, 2^2)

```python
d=np.random.default_rng(0).normal(5,2,10000); print(round(d.mean(),2), round(d.std(),2))
```

### E2 — 68-95-99.7

```python
d=np.random.default_rng(1).normal(0,1,100000)
print([round(np.mean(np.abs(d)<k),3) for k in (1,2,3)])
```

### E3 — binomial mean

```python
k=np.random.default_rng(2).binomial(20,0.5,10000); print(round(k.mean(),2))   # ~10
```

### E4 — pdf peak

```python
print(round(normal_pdf(2,2,0.5),4), round(1/(0.5*np.sqrt(2*np.pi)),4))
```

### E5 — 2-D correlated Gaussian

```python
rng=np.random.default_rng(3); S=np.array([[1,0.8],[0.8,1]])
X=rng.multivariate_normal([0,0],S,5000)
print(np.cov(X.T).round(2))
print(np.linalg.eigh(np.cov(X.T))[1][:,-1].round(3))   # ~[.707,.707]
```
