# Inference & hypothesis testing — practice (worked & executed)

```python
import numpy as np
from scipy import stats
Phi=stats.norm.cdf
nA=nB=2000; cA,cB=200,240; pA,pB=cA/nA,cB/nB; p=(cA+cB)/(nA+nB)
se=np.sqrt(p*(1-p)*(1/nA+1/nB)); z=(pB-pA)/se; pval=2*(1-Phi(abs(z)))
print("lift",round(pB-pA,3),"z",round(z,2),"p",round(pval,4))
```

### E1 — confidence interval

```python
se=10/np.sqrt(400); print([round(50-1.96*se,3), round(50+1.96*se,3)])   # [49.02, 50.98]
```

### E2 — A/B z & p

```python
print("z",round(z,2),"p",round(pval,4))
```

### E3 — two-sided p from CDF

```python
print(round(2*(1-Phi(2.02)),4))
```

### E4 — false positive rate (~5%)

```python
rng=np.random.default_rng(0); n=1000; N=10000
cA=rng.binomial(n,0.1,N); cB=rng.binomial(n,0.1,N); pool=(cA+cB)/(2*n)
se=np.sqrt(pool*(1-pool)*(2/n)); z=np.divide(cB/n-cA/n, se, out=np.zeros(N), where=se>0)
print("false positive rate ~", round(np.mean(2*(1-Phi(np.abs(z)))<0.05),3))
```

### E5 — power vs n

```python
def power(nn,pa=0.10,pb=0.12,N=5000):
    cA=rng.binomial(nn,pa,N); cB=rng.binomial(nn,pb,N); pool=(cA+cB)/(2*nn)
    se=np.sqrt(pool*(1-pool)*(2/nn)); z=np.divide(cB/nn-cA/nn, se, out=np.zeros(N), where=se>0)
    return np.mean(2*(1-Phi(np.abs(z)))<0.05)
print("power n=1000", round(power(1000),3), "| n=5000", round(power(5000),3))
```
