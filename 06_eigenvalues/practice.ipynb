# Eigenvalues & eigenvectors — practice (worked & executed)

```python
import numpy as np
A=np.array([[2,1],[1,2.]]); vals,vecs=np.linalg.eig(A)
print("vals",vals,"| Av==lv:",np.allclose(A@vecs[:,0],vals[0]*vecs[:,0]))
print("sum=trace",np.isclose(vals.sum(),np.trace(A)),"prod=det",np.isclose(vals.prod(),np.linalg.det(A)))
```

### E1 — diagonal by inspection

```python
print(np.linalg.eigvals([[3,0],[0,1]]))   # 3, 1
```

### E2 — characteristic equation

```python
print(np.roots([1,-4,3]))   # lambda^2-4l+3 -> 3, 1
```

### E3 — sum/product checks (random 3x3)

```python
rng=np.random.default_rng(1); M=rng.random((3,3)); v=np.linalg.eigvals(M)
print(np.isclose(v.sum().real,np.trace(M)), np.isclose(v.prod().real,np.linalg.det(M)))
```

### E4 — rotation -> complex eigenvalues

```python
print(np.linalg.eigvals([[0,-1],[1,0]]))   # +/- i
```

### E5 — power iteration -> dominant eigenvector

```python
A=np.array([[2.,1],[1,2]]); x=np.random.default_rng(0).random(2)
for _ in range(60): x=A@x; x=x/np.linalg.norm(x)
print(np.round(x,3))   # ~[0.707,0.707]
```
