# Eigenvalues & eigenvectors — practice

Read `theory.html` first. Do the 2x2s by hand, then verify. Remember: `eig` returns eigenvectors as COLUMNS.

```python
import numpy as np

A = np.array([[2., 1.], [1., 2.]])
vals, vecs = np.linalg.eig(A)
print("eigenvalues :", vals)
print("eigenvectors (cols):\n", vecs)
print("check Av == lambda v:", np.allclose(A @ vecs[:,0], vals[0]*vecs[:,0]))
print("sum==trace:", np.isclose(vals.sum(), np.trace(A)), " prod==det:", np.isclose(vals.prod(), np.linalg.det(A)))
```
