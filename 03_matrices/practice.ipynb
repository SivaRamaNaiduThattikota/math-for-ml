# Matrices as transformations — practice (worked & executed)

```python
import numpy as np
A=np.array([[2,1],[0,3.]]); x=np.array([1,1.])
print("A@x =", A@x, "| det =", np.linalg.det(A))
print("cols:", A@np.array([1,0]), A@np.array([0,1]))
```

### E1 — scaling

```python
print(np.array([[3,0],[0,1]])@np.array([2,5]))   # [6 5]
```

### E2 — read basis images (shear)

```python
S=np.array([[1,1],[0,1]]); print(S@np.array([1,0]), S@np.array([0,1]))
```

### E3 — 90 degree rotation

```python
R=np.array([[0,-1],[1,0]]); print(R@np.array([1,0]))   # [0 1]
```

### E4 — singular matrix

```python
print(np.linalg.det([[2,1],[4,2]]))   # 0
```

### E5 — idempotent projection

```python
P=np.array([[1,0],[0,0]]); print(P@P, "| P^2==P:", np.allclose(P@P,P))
```
