# Matrix multiplication — practice

Read `theory.html` first. Compute each product by hand before running. Remember: `@` is matmul, `*` is element-wise.

```python
import numpy as np

A = np.array([[1., 2.], [0., 1.]])
B = np.array([[2., 0.], [1., 3.]])
print("A @ B =\n", A @ B)          # 'B then A'
print("B @ A =\n", B @ A)          # different!
print("commute?", np.allclose(A @ B, B @ A))
print("element-wise A*B (NOT matmul) =\n", A * B)
```
