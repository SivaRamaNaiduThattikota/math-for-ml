# Bayesian inference & conjugate priors — practice

Read `theory.html` first. Beta(a,b) + h heads, t tails -> Beta(a+h, b+t).

```python
import numpy as np
from scipy import stats

a0,b0 = 2,2; h,t = 14,6
a,b = a0+h, b0+t
print('posterior Beta(%d,%d) mean=%.3f' % (a,b,a/(a+b)))          # 16 8 0.667
print('95%% credible:', np.round(stats.beta.ppf([.025,.975],a,b),3))  # [0.471 0.836]
print('MLE, MAP, mean:', h/(h+t), (a-1)/(a+b-2), a/(a+b))         # 0.70 0.682 0.667
```
