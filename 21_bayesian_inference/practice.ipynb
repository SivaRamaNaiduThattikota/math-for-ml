# Bayesian inference & conjugate priors — practice (worked & executed)

```python
import numpy as np
from scipy import stats
a0,b0=2,2; h,t=14,6; a,b=a0+h,b0+t
print("posterior Beta(%d,%d) mean=%.3f" % (a,b,a/(a+b)))
print("95%% credible:", np.round(stats.beta.ppf([.025,.975],a,b),3))
```

### E1 — posterior params & mean

```python
print(a,b,round(a/(a+b),3))   # 16 8 0.667
```

### E2 — credible interval

```python
print(np.round(stats.beta.ppf([.025,.975],16,8),3))   # [0.471 0.836]
```

### E3 — MLE vs MAP vs mean

```python
print("MLE",h/(h+t),"MAP",round((a-1)/(a+b-2),3),"mean",round(a/(a+b),3))
```

### E4 — prior washes out

```python
def mean_after(h,t,a0=2,b0=2): return (a0+h)/(a0+b0+h+t)
print("4 flips (3H1T):", round(mean_after(3,1),3), "| 400 flips (300H100T):", round(mean_after(300,100),3))
```

### E5 — Thompson sampling

```python
rng=np.random.default_rng(0)
A=stats.beta(30,70); B=stats.beta(45,55)   # B better (~0.45 vs 0.30)
picks=["B" if B.rvs(random_state=rng)>A.rvs(random_state=rng) else "A" for _ in range(10000)]
print("fraction picking B:", round(picks.count("B")/10000,3))
```
