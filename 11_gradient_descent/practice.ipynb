# Gradient descent — practice

Read `theory.html` first. The whole algorithm is `w -= eta * grad(w)` in a loop.

```python
import numpy as np

grad = lambda w: 2*(w - 3)      # L(w) = (w-3)^2, min at w=3
w, eta = 0.0, 0.1
for _ in range(40):
    w -= eta * grad(w)
print(w)   # ~3.0
```
