# Gradient descent — practice (worked & executed)

```python
import numpy as np
g=lambda w:2*(w-3); w=0.0
for _ in range(40): w-=0.1*g(w)
print("converged w =", round(w,4))
```

### E1 — converge

```python
w=0.0
for _ in range(40): w-=0.1*g(w)
print(round(w,4))
```

### E2 — diverge

```python
w=0.0
for _ in range(20): w-=1.1*g(w)
print("eta=1.1 ->", w)
```

### E3 — linear regression by GD

```python
rng=np.random.default_rng(0); X=rng.random(200); y=2.5*X+0.1*rng.standard_normal(200); w=0.0
for _ in range(300): w-=0.5*np.mean(2*X*(w*X-y))
print(round(w,3))   # ~2.5
```

### E4 — learning-rate sweep

```python
for eta in [0.01,0.1,0.5,1.0,1.1]:
    w=0.0
    for _ in range(30): w-=eta*g(w)
    print(eta, round(w,3))
```

### E5 — momentum

```python
w=v=0.0
for _ in range(40): v=0.9*v-0.1*g(w); w+=v
print("with momentum w =", round(w,3))
```
