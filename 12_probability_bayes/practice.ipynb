# Probability & Bayes — practice

Read `theory.html` first. Always fold in the base rate.

```python
import numpy as np

prev, sens, spec = 0.01, 0.99, 0.95
p_pos = sens*prev + (1-spec)*(1-prev)
print('P(D | +) =', round(sens*prev / p_pos, 3))   # 0.167

# posterior ∝ likelihood × prior
prior = np.array([0.01, 0.99]); like = np.array([sens, 1-spec])
print((prior*like / (prior*like).sum()).round(3))   # [0.167 0.833]
```
