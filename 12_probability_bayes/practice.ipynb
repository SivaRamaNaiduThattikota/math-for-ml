# Probability & Bayes — practice (worked & executed)

```python
import numpy as np
prev,sens,spec=0.01,0.99,0.95
print("P(D|+) =", round(sens*prev/(sens*prev+(1-spec)*(1-prev)),3))   # 0.167
prior=np.array([0.01,0.99]); like=np.array([sens,1-spec])
print((prior*like/(prior*like).sum()).round(3))   # [0.167 0.833]
```
