# SVD, PCA & low-rank — practice (worked & executed)

```python
import numpy as np
A=np.array([[3,1],[1,2],[0,1.]]); U,S,Vt=np.linalg.svd(A,full_matrices=False)
print("S",S.round(4),"| reconstruct",np.allclose(A,U@np.diag(S)@Vt))
```

### E1 — diagonal SVD

```python
print(np.linalg.svd([[3,0],[0,2.]],compute_uv=False))   # [3,2]
```

### E2 — sigma = sqrt(eig(A^T A))

```python
rng=np.random.default_rng(2); A=rng.random((3,2)); S=np.linalg.svd(A,compute_uv=False)
lam=np.sort(np.linalg.eigvals(A.T@A).real)[::-1]; print(np.allclose(S,np.sqrt(lam)))
```

### E3 — rank-1 approximation

```python
rng=np.random.default_rng(3); A=rng.random((4,4)); U,S,Vt=np.linalg.svd(A)
A1=S[0]*np.outer(U[:,0],Vt[0])
print("full err", round(np.linalg.norm(A-U@np.diag(S)@Vt),6), "| rank1 err", round(np.linalg.norm(A-A1),3))
```

### E4 — PCA finds the spread direction

```python
rng=np.random.default_rng(4); t=rng.standard_normal(500)
X=np.column_stack([t,t+0.1*rng.standard_normal(500)]); Xc=X-X.mean(0)
_,S,Vt=np.linalg.svd(Xc,full_matrices=False); print(np.round(Vt[0],3))   # ~[.707,.707]
```

### E5 — explained variance

```python
rng=np.random.default_rng(5); X=rng.random((200,5)); Xc=X-X.mean(0)
S=np.linalg.svd(Xc,compute_uv=False); ratio=S**2/np.sum(S**2)
print("cumulative:", np.cumsum(ratio).round(3))
```
