# SVD, PCA & low-rank — practice

Read `theory.html` first. Note: `np.linalg.svd` returns U, S (1-D singular values), Vt (= Vᵀ).

```python
import numpy as np

A = np.array([[3., 1.], [1., 2.], [0., 1.]])   # 3x2, rectangular
U, S, Vt = np.linalg.svd(A, full_matrices=False)
print("singular values:", S)
print("reconstruct ok :", np.allclose(A, U @ np.diag(S) @ Vt))
print("sigma == sqrt(eig(A^T A)):",
      np.allclose(S, np.sqrt(np.sort(np.linalg.eigvals(A.T @ A).real)[::-1])))
```
