# Sampling & the CLT — practice (worked & executed)

```python
import numpy as np
rng=np.random.default_rng(0)
means=[rng.exponential(1.0,30).mean() for _ in range(5000)]
print(round(np.mean(means),3), round(np.std(means),3))   # ~1.0, ~0.183
```

### E1 — uniform means

```python
m=[rng.random(30).mean() for _ in range(5000)]
print(round(np.mean(m),3), round(np.std(m),4), "vs", round((1/np.sqrt(12))/np.sqrt(30),4))
```

### E2 — sqrt(n) law

```python
d=rng.standard_normal(100000)
se100=d.std()/np.sqrt(100); se400=d.std()/np.sqrt(400)
print(round(se100,4), round(se400,4), "ratio", round(se100/se400,3))   # ~2
```

### E3 — raw vs means (exponential)

```python
raw=rng.exponential(1,5000); means=np.array([rng.exponential(1,50).mean() for _ in range(5000)])
print("raw skew-ish std", round(raw.std(),3), "| means std", round(means.std(),3))
```

### E4 — bootstrap SE of the median

```python
d=rng.exponential(1,500)
boot=[np.median(rng.choice(d,500,replace=True)) for _ in range(2000)]
print("bootstrap SE(median) =", round(np.std(boot),4))
```

### E5 — ensembling variance reduction

```python
for k in (1,4,16):
    err=rng.standard_normal((100000,k)).mean(1)
    print(k, round(err.std(),3))   # ~1, ~0.5, ~0.25
```
