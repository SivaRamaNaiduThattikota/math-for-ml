# Sampling & the CLT — practice

Read `theory.html` first. The CLT normalizes the *mean*, not the raw data.

```python
import numpy as np

# means of a skewed source look Gaussian
means = [np.random.exponential(1.0, 30).mean() for _ in range(5000)]
print(round(np.mean(means),3), round(np.std(means),3))   # ~1.0, ~1/sqrt(30)=0.183

# bootstrap SE of the mean
data = np.random.exponential(1.0, 1000)
boot = [np.random.choice(data, 1000, replace=True).mean() for _ in range(2000)]
print(round(np.std(boot),4), round(data.std(ddof=1)/np.sqrt(len(data)),4))  # match
```
