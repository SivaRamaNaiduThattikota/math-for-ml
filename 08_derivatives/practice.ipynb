# Derivatives & gradients — practice (worked & executed)

```python
import numpy as np
def deriv(f,x,h=1e-5): return (f(x+h)-f(x-h))/(2*h)
print(deriv(lambda x:x**2,3.0), deriv(np.sin,0.0), round(deriv(np.exp,1.0),4))
```

### E1 — polynomial

```python
f=lambda x:3*x**2+2*x-5; print("numeric", round(deriv(f,2),4), "| exact 6x+2 =", 6*2+2)
```

### E2 — critical point of x^2

```python
print("slope at 0:", deriv(lambda x:x**2,0.0), "-> minimum")
```

### E3 — critical points of x^3 - 3x

```python
f=lambda x:x**3-3*x
for x in [-1,1]: print("x=",x,"slope",round(deriv(f,x),3))
```

### E4 — deriv(sin) = cos

```python
xs=np.linspace(0,2*np.pi,100); print(np.allclose(deriv(np.sin,xs),np.cos(xs),atol=1e-4))
```

### E5 — one gradient-descent step

```python
L=lambda w:(2*w-6)**2; g=lambda w:8*w-24
w=0; w=w-0.1*g(w); print("after 1 step w =", w, "(min at 3)")
```
