# Derivatives & gradients — practice

Read `theory.html` first. Do the by-hand derivatives, then confirm with finite differences.

```python
import numpy as np

def deriv(f, x, h=1e-5):
    return (f(x + h) - f(x - h)) / (2*h)

print(deriv(lambda x: x**2, 3.0))   # 6.0  (2x)
print(deriv(np.sin, 0.0))           # 1.0  (cos 0)
print(deriv(np.exp, 1.0))           # e ~ 2.718
```
