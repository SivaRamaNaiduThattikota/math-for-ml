# Statistical tests: t, chi2 & ANOVA — practice (worked & executed)

```python
import numpy as np
from scipy import stats
rng=np.random.default_rng(1)
a=rng.normal(100,15,30); b=rng.normal(108,15,30)
print("two-sample t:", np.round(stats.ttest_ind(a,b),4))
```

### E1 — two-sample t

```python
print(np.round(stats.ttest_ind(a,b),4))   # t=-2.505 p=0.0151
```

### E2 — chi2 goodness of fit

```python
print(np.round(stats.chisquare([16,18,16,14,22,14]),4))   # 2.72 0.7431
```

### E3 — chi2 independence

```python
print(np.round(stats.chi2_contingency([[30,10],[20,40]])[:2],5))   # 15.042 0.00011
```

### E4 — ANOVA

```python
g1=rng.normal(50,5,20); g2=rng.normal(52,5,20); g3=rng.normal(58,5,20)
print(np.round(stats.f_oneway(g1,g2,g3),4))
```

### E5 — t -> normal

```python
print([round(stats.t.pdf(0,df),4) for df in (1,5,30)], round(stats.norm.pdf(0),4))
```
