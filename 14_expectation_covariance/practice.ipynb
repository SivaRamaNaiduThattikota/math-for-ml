# Expectation, variance & covariance — practice

Read `theory.html` first. Note np.cov uses ddof=1 and rowvar=True by default.

```python
import numpy as np

x = np.array([2,4,4,4,5,5,7,9.])
print(x.mean(), x.var(), x.std())            # 5.0, 4.0, 2.0
print((x**2).mean() - x.mean()**2)           # 4.0  (E[X^2]-E[X]^2)

a = np.random.normal(size=2000)
b = 0.7*a + 0.3*np.random.normal(size=2000)
print(np.cov(a, b).round(2))                 # covariance matrix
print(np.corrcoef(a, b)[0,1].round(2))       # correlation
```
