# Expectation, variance & covariance — practice (worked & executed)

```python
import numpy as np
x=np.array([2,4,4,4,5,5,7,9.])
print(x.mean(), x.var(), x.std())
print("E[X^2]-E[X]^2 =", (x**2).mean()-x.mean()**2)
```

### E1 — die mean & variance

```python
r=np.random.default_rng(0).integers(1,7,100000); print(round(r.mean(),3), round(r.var(),3))  # ~3.5, ~2.92
```

### E2 — Var = E[X^2]-E[X]^2

```python
x=np.random.default_rng(1).random(1000); print(np.isclose(x.var(),(x**2).mean()-x.mean()**2))
```

### E3 — cov vs corr under scaling

```python
rng=np.random.default_rng(2); a=rng.standard_normal(2000); b=2*a+rng.standard_normal(2000)
print("cov", round(np.cov(a,b)[0,1],3), "corr", round(np.corrcoef(a,b)[0,1],3))
a2=1000*a
print("cov scaled", round(np.cov(a2,b)[0,1],1), "corr scaled", round(np.corrcoef(a2,b)[0,1],3))
```

### E4 — averaging reduces variance

```python
rng=np.random.default_rng(3)
print("single var", round(rng.standard_normal(100000).var(),3),
      "| mean-of-25 var", round(rng.standard_normal((100000,25)).mean(1).var(),3))
```

### E5 — covariance eigenvector = main axis

```python
rng=np.random.default_rng(4); a=rng.standard_normal(2000); b=0.9*a+0.4*rng.standard_normal(2000)
w,V=np.linalg.eigh(np.cov(np.vstack([a,b]))); print("top eigenvector", V[:,-1].round(3))
```
